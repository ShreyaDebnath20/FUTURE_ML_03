# AI Resume Screening - Results

## Model Performance

- Model Used: Random Forest Classifier
- Accuracy Achieved: 100%

## Important Features

Based on feature importance analysis:

1. AI Score (0-100)
2. Experience (Years)
3. Projects Count
4. Salary Expectation ($)

## Key Findings

- AI Score is the strongest predictor of recruiter decisions.
- Candidates with more experience have a higher chance of selection.
- Project count positively influences recruiter decisions.
- Salary expectation has comparatively lower impact.

## Conclusion

This project successfully predicts recruiter decisions using machine learning techniques. The Random Forest model achieved excellent performance on the dataset and identified AI Score, Experience, and Projects Count as the most influential factors.
